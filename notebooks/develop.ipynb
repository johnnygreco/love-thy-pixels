```python
from astropy import units as u
from photutils import SkyCircularAperture
from astropy.coordinates import SkyCoord
from photutils import aperture_photometry
```

```python
coord = SkyCoord(219.9580, -0.7127, unit='deg')
```

```python
ap = SkyCircularAperture(coord, 1 * u.arcsec)
```

```python
aperture_photometry(galaxy.red_data, ap, wcs=galaxy.wcs)['aperture_sum'] / aperture_photometry(galaxy.blue_data, ap, wcs=galaxy.wcs)['aperture_sum']
```
