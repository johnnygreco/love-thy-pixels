**[Digital images](https://en.wikipedia.org/wiki/Digital_image)** are *everywhere* $-$ how many do you think you have seen and interacted with so far today? Much more than pretty pictures and weekend entertainment (movies are lots of digital images stitched together), digital images and [imaging processing](https://en.wikipedia.org/wiki/Digital_image_processing) are routinely used by scientist and engineers in fields ranging from astronomy to medicine to national defense. But what is a digital image exactly? 

A Digital image is a two-dimensional grid of numbers (also known as an array), which represent *intensity* values to be displayed on a monitor. Each element of the grid is called a "picture element" or [**pixel**](https://en.wikipedia.org/wiki/Pixel) (hence the name of our project... **[Love thy Pixels](https://github.com/johnnygreco/love-thy-pixels)**). More pixels = higher resolution = more information recorded, which is why camera makers like Canon and Nikon love to boast about the number of *megapixels* (mega = $10^6$ = 1 million) in their cameras.

```python
print(circle)
```

Notice there are only two numbers `0` and `1`. This is known as a [*binary* image](https://en.wikipedia.org/wiki/Binary_image).

A [grayscale](https://en.wikipedia.org/wiki/Grayscale) image uses more numbers to represent different shades of gray. An 8-[bit](https://en.wikipedia.org/wiki/Bit) image has $2^8 = 256$ different gray tones.

## Section 1: Color images and more cute animals

Color images are created by mixing [red, green, and blue](https://en.wikipedia.org/wiki/RGB_color_model) (RGB) [channels](https://en.wikipedia.org/wiki/Channel_(digital_image)) together. Often (like in your phone) these colored images are taken simultaneously using filters that only allow red, green, or blue light to hit a detector. An 8-bit RGB image is composed of three images (one for each channel), where the image pixels store intensity value between 0 and 255 (256 unique values). 

In this section, we'll learn how to display and manipulate color images with Python. As our test subjects, we will use adorable animal pictures!

## Section 2: Images as Arrays

Remember that to a computer, an image is really just a list of numbers called an *array*. Each number in the array has an *index* and a *value*. The index tells you where in the array that particular number (pixel) is located, and the value determines how that part of the image will appear when you display it.
