## Section 2: Stellar Populations

*Massive* stars are the rock stars of the universe. They live fast and die *young*. Their surfaces are much hotter than our Sun, leading to much bluer colors. Therefore, when a cluster of stars (or a galaxy) appears very blue, we know it contains young stars. Sweeping many subtleties under the cosmic rug, this means the color of a galaxy tells us the average age of its stars. 

The three color images below show simulated star clusters composed of stars with a range of masses, which were born 300 million, 3 billion, and 10 billion years ago. Note the clusters grow redder with age. **Remember:** In astronomy, 300 millions years is a blink of an eye!

![simulated globular clusters](https://raw.githubusercontent.com/johnnygreco/love-thy-pixels/master/images/gc-ages.png)

In this section, we will measure the colors of galaxies and estimate the average age of their stars. We will measure the light in red, green, and blue filters using *aperture photometry*, which is basically just adding up all the pixels in a particular region of the image. We will use circular regions.

Let's start by loading galaxy number 1:

```python
galaxy = AstroImage(1) # insert galaxy number into the function
```

Display a RGB image of the galaxy:

```python
galaxy.display_rgb_image()
```

 ---
 
 Use [JS9](https://johnnygreco.space) to find the coordiates of different regions in the galaxy using a circular aperture. 
 
 
Create a list of positions in this notebook like this: `positions = [(x1, y1), (x2, y2), (x3, y3), ...]`

```python
positions = [(1297.50, 919.50), (1077.50, 1095.50)]
```

To measure the light in the red, green, and blue filters, use the `galaxy.measure_light_in_circle(positions, radius)` function. The first argument to the function is the list of positions you just measured, and the second argument is the radius of the circular aperture in units of pixels. Start with a circular aperture with a radius of 15 pixels. Then try different aperture sizes tuned to the structures in the galaxy you are trying to measure. 

```python
red, green, blue = galaxy.measure_light_in_circle(positions, 15)# enter the positions and radius into the function
```

The function returns the arrays `red`, `green`, and `blue`. The elements of these arrays correspond to the amount of light in each filter at the positions in your list.

In astronomy, we define color as the ratio of light in two filters. Calculate the galaxy's color by taking the ratio of the red and blue arrays: `red / blue`.

```python
red / blue
```

You can plot your apertures on the single-band images using the optional `positions` and `radius` arguments to `galaxy.display_single_band_image`. 
The numbers next to the apertures correspond to the order of your `positions` array.

```python
galaxy.display_single_band_image('red', positions=positions, radius=15)
```
